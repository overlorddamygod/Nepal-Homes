```python
df.drop(["built_year","state","is_price_on_call","property_face"],axis=1, inplace=True)
```

```python
df.loc[df.district.isnull(),"district"] = df["district"].value_counts().index[0]
```

```python
median_by_district = df.groupby("district").median()["amount"]
df.loc[df.amount.isnull(),"amount"] = df.apply(lambda x:median_by_district[x["district"]],axis=1)
df["amount"] = df.amount / 1000000
```

```python
impute_cols = ["bedroom","bathroom","road_access","area_sq_ft","floors"]
for col in impute_cols:
    df[col] = df[col].fillna(df[col].median())
```

# Outlier Removal

```python
def outlier_removal(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1
    iqr_dist = 3.5
    return df[(df[col] > q1 - iqr_dist*iqr) & (df[col] < q3+iqr_dist*iqr)]
```

```python
for col in df.select_dtypes(include=np.number).columns:
    if not col in ["floors","bathroom"]:
        df = outlier_removal(df,col)
df["is_bungalow"] = df.title.apply(lambda x: 1 if "bungalow" in x.lower() else 0)
df = df[df.bathroom < 20]
```
