```python
df.isna().sum()/ df.shape[0] * 100 > 10
```

```python
missing_df = df.isna()
```

```python
missing_cols = ["built_year","floors","bedroom","bathroom","road_access","property_face"]

fig, axes = plt.subplots(2,3,figsize=(12,6))

for i,col in enumerate(missing_cols):
    sns.countplot(missing_df[col],ax=axes[int(i/3),int(i % 3)])
```

```python
fig, axes = plt.subplots(2,3,figsize=(12,12))

for i,col in enumerate(missing_cols):
    sns.boxplot(x=missing_df[col],y=df.amount,ax=axes[int(i/3),int(i % 3)])
```
